```python
import numpy as np 
import pandas as pd 

data_train = pd.read_csv('../input/titanic/train.csv')
data_test = pd.read_csv('../input/titanic/test.csv')
cols =['Pclass','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked','Survived'] 
X_train = data_train[cols]
#y_train = data_train['Survived']
```

```python
drop_X_train = X_train.drop(['Cabin'],axis=1)
```

```python
drop_X_train.dropna(axis=0,inplace=True)
```

```python
drop_X_train
```

```python
num_cols = drop_X_train.select_dtypes(exclude ='object')
cat_cols = drop_X_train.select_dtypes(include ='object')
```

```python
num_cols
```

```python
cat_cols
```

```python
good_cat_cols = [co for co in cat_cols if(cat_cols[co].nunique() < 10)]
```

```python
cat_cols
```

```python
good_cat_cols
```

```python
good_cat_cols = cat_cols[good_cat_cols]
```

```python
drop_X_train
```

```python
num_cols.drop('Age',axis=1,inplace=True)
```

```python
num_cols
```

```python
num_cols['Fare'].isnull().sum()
```

```python
good_cat_cols
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
for x in good_cat_cols:
    good_cat_cols[x] = lb.fit_transform(good_cat_cols[x])
```

```python
good_cat_cols
```

```python
good_cat_cols[num_cols.columns] = num_cols
```

```python
final_X = good_cat_cols
y = final_X['Survived']
final_X.drop('Survived',axis=1,inplace=True)
```

```python
y
```

```python
final_X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_valid,y_train,y_valid = train_test_split(final_X,y,train_size=0.80,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(n_estimators=300,learning_rate=0.1)
```

```python
model.fit(X_train,y_train)
```

```python
pre = model.predict(X_valid)
```

```python
from sklearn.metrics import accuracy_score
sc = accuracy_score(pre,y_valid)
```
